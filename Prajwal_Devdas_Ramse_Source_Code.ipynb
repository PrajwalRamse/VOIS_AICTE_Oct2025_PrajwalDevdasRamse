```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv('/content/1730285881-Airbnb_Open_Data_1.csv',low_memory=False)
```

```python
df['neighbourhood group']=df['neighbourhood group'].replace('brookln','Brooklyn')
```

```python
df[df['neighbourhood group']=='brookln']
```

```python
# Drop duplicate records
df.drop_duplicates(inplace = True)
```

```python
df.drop(columns=['house_rules', 'license'], axis=1, inplace=True, errors='ignore')
```

```python
if 'price' in df.columns:
    df['price'] = df['price'].astype(str).str.replace('$', '', regex=False)
if 'service_fee' in df.columns:
    df['service_fee'] = df['service_fee'].astype(str).str.replace('$', '', regex=False)
```

```python
if 'price' in df.columns:
    df['price'] = df['price'].str.replace(',', '', regex=False)
if 'service_fee' in df.columns:
    df['service_fee'] = df['service_fee'].str.replace(',', '', regex=False)
```

```python
df.rename(columns={
    'price': 'price_$',
    'service_fee': 'service_fee_$'
}, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
# Convert to float (for price & service fee)
for col in ['price_$', 'service_fee_$']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert to int (for host_id & Construction year)
if 'host_id' in df.columns:
    df['host_id'] = pd.to_numeric(df['host_id'], errors='coerce').astype('Int64')

if 'Construction year' in df.columns:
    df['Construction year'] = pd.to_numeric(df['Construction year'], errors='coerce').astype('Int64')

# Convert to datetime (for host_since & last_review)
for date_col in ['host_since', 'last_review']:
    if date_col in df.columns:
        df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
```

```python
if 'availability_365' in df.columns:
    df.drop(df[df['availability_365'] > 500].index, inplace=True)
```

```python
# Which neighborhood group has the highest number of listings?
hood_group = df['neighbourhood group'].value_counts().to_frame()
hood_group
```

```python
# Which neighborhood group has the highest number of listings?
hood_group_bar = plt.bar(hood_group.index, hood_group.loc[:, 'count'])
plt.bar_label(hood_group_bar, labels = hood_group.loc[:, 'count'], padding = 4)
plt.ylim(0, 40000)
plt.xlabel('Neighborhood Groups')
plt.ylabel('Number of Listings')
plt.xticks(rotation = 45)
plt.title('Which Neighborhood Groups has the highest number of Listings?')
```

```python
# Calculate average price by neighborhood group
avg_price = df.groupby('neighbourhood group')['price_$'].mean().sort_values(ascending=False)

# Select top 5 and round values
top_5_avg_price = avg_price.head(5).round(2)

# Plot bar chart
plt.figure(figsize=(10, 7))
plt.bar(top_5_avg_price.index, top_5_avg_price.values, edgecolor='black')
plt.title('Top 5 Average Price Listings ($)')
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Price ($)')
plt.show()
```

```python
# --- 1) detect price column (tries several common names) ---
possible_price_cols = ['price ($)', 'price_$', 'price']
price_cols_found = [c for c in possible_price_cols if c in df.columns]
if not price_cols_found:
    raise KeyError("No price column found. Expected one of: " + ", ".join(possible_price_cols) +
                   ". Columns available: " + ", ".join(df.columns))
price_col = price_cols_found[0]

# --- 2) check for Construction year column ---
if 'Construction year' not in df.columns:
    raise KeyError("'Construction year' column not found. Columns available: " + ", ".join(df.columns))

# --- 3) clean and convert price to numeric (remove $ and commas if present) ---
df[price_col] = pd.to_numeric(df[price_col].astype(str).str.replace(r'[\$,]', '', regex=True), errors='coerce')

# --- 4) convert Construction year to numeric (coerce bad values to NaN) ---
df['Construction year'] = pd.to_numeric(df['Construction year'], errors='coerce')

# --- 5) drop rows missing either price or construction year ---
tmp = df.dropna(subset=[price_col, 'Construction year']).copy()

# --- 6) group and compute mean price per construction year ---
avg_price_by_year = tmp.groupby('Construction year')[price_col].mean().sort_index()
counts_by_year = tmp.groupby('Construction year')[price_col].count()

# --- 7) optional: remove years with very few listings to reduce noise (adjust min_count as needed) ---
min_count = 5
valid_years = counts_by_year[counts_by_year >= min_count].index
avg_price_by_year = avg_price_by_year.loc[valid_years]

# --- 8) Plotting ---
plt.figure(figsize=(12,6))
plt.plot(avg_price_by_year.index, avg_price_by_year.values)
plt.scatter(avg_price_by_year.index, avg_price_by_year.values)   # markers to show points
plt.xlabel('Construction Year')
plt.ylabel(f'Average Price ({price_col})')
plt.title('Average Price by Construction Year')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# adjust names as per dataset
hosts = (
    df.groupby('host name')['calculated host listings count']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .to_frame()
)

# Plot bar chart
plt.figure(figsize=(12, 6))
plt.bar(hosts.index, hosts.iloc[:, 0], color='skyblue', edgecolor='black')

plt.xlabel('Host Name')
plt.ylabel('Calculated Host Listings Count')
plt.title('Top 10 Hosts by Calculated Host Listings Count')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, hosts.iloc[:, 0].max() * 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
